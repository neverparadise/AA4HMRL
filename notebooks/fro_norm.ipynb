```python
LOG_STD_MIN = -5
LOG_STD_MAX = 1
```

```python
a_mu = torch.zeros([32, 6])
a_logstd = torch.ones([32, 6])
```

```python
a_logstd = LOG_STD_MIN + 0.5 * (LOG_STD_MAX - LOG_STD_MIN) * (a_logstd + 1)
```

```python
a_logstd
```

```python
from torch.distributions import Normal
```

```python
a_std = a_logstd.exp()
```

```python
a_std
```

```python
dist = Normal(a_mu, a_std)
```

```python
entropy = dist.entropy().sum(-1)
```
